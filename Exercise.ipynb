```python
from PIL import Image
from scipy.stats import norm
```

```python
!wget https://github.com/XSPZ/machine_learning_examples/blob/master/svsss.png?raw=true
```

```python
im = Image.open('svsss.png?raw=true')
```

```python
gray = np.mean(im, axis=2)
```

```python
x = np.linspace(-6, 6, 50)
fx = norm.pdf(x, loc=0, scale=1)
```

```python
filt = np.outer(fx, fx)
```

```python
plt.imshow(filt, cmap='gray');
```

```python
from scipy.signal import convolve2d
```

```python
out = convolve2d(gray, filt)
```

```python
plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(out, cmap='gray');
```

```python
from numpy.ma.core import size
Hx = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])
Hx.reshape(3, 3)
Hx[0,2]
```

```python
Hy = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
Hy.reshape(3, 3)
Hy[0,1]
```

```python
gray = np.mean(im, axis=2)
gray
```

```python
Gx = convolve2d(gray, Hx)
```

```python
Gy = convolve2d(gray, Hy)
```

```python
G = np.sqrt(Gx ** 2 + Gy ** 2)
G
```

```python
plt.imshow(G, cmap='cool')
```
